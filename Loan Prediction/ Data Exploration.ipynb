```python
data_category_column = dataf.select_dtypes(exclude=np.number).columns
```

```python
data_integer_column = dataf.columns.difference(data_category_column)
```

```python
dataf[data_category_column].nunique()
```

```python
dataf_onehot_column = pd.get_dummies(dataf[data_category_column].drop(columns=["Loan_ID","Loan_Status"]))
```

```python
data_model = pd.concat([dataf["Loan_ID"],dataf["Loan_Status"],dataf_onehot_column,dataf[data_integer_column]],axis=1)
```

```python
data_model.head()
```

```python
data_model.loc[data_model["Loan_Status"].notnull(),"source"]="Train"
```

```python
data_model["source"].fillna("Test",inplace=True)
```

```python
train = data_model.loc[data_model['source']=="Train"]
test = data_model.loc[data_model['source']=="Test"]

#Drop unnecessary columns:
test.drop(['source'],axis=1,inplace=True)
train.drop(['source'],axis=1,inplace=True)

#Export files as modified versions:
train.to_csv("train_modified.csv",index=False)
test.to_csv("test_modified.csv",index=False)
```
